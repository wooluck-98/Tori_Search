```python
import copy

# 행렬을 출력하는 함수
def show_arr(input_arr):
    for i in range(y_size):
        for j in range(x_size):
            if (input_arr[i][j] == True):
                input_arr[i][j] = 1
            else:
                input_arr[i][j] = 0                
    for i in range(y_size-1, -1, -1):
        print(input_arr[i])
        
if __name__ == '__main__':
    # x축의 크기, y축의 크기, 문제 입력
    x_size = int(input("x축의 크기 :"))
    y_size = int(input("y축의 크기 :"))
    arr = []
    for i in range(y_size):
        temp = list(input())
        for j in range(x_size):
            if temp[j] == '1':
                temp[j] = True
                continue
            temp[j] = False
        arr.append(temp)

    dx = [0, 0, 0, -1, 1]
    dy = [0, 1, -1, 0 ,0]
    
    # 비트마스크 사용
    for bit in range(1<<x_size):
        temp_arr = []
        temp_arr = copy.deepcopy(arr)
        root = []
        arr_shape=[]
        # 중간과정을 저장
        print_arr = copy.deepcopy(temp_arr)
        arr_shape.append(print_arr)
    
        # 첫번째 줄에 대한 시행
        for i in range(x_size):
            if bit & (1<<i):
                # 카드를 클릭
                for k in range(5):
                    nx = i + dx[k]
                    ny = 0 + dy[k]
                    if 0 <= nx and nx < x_size and 0 <= ny and ny < y_size:
                        temp_arr[ny][nx] = not temp_arr[ny][nx]
                # 카드를 클릭한 좌표와 변화된 리스트 저장
                root.append((i+1, 0+1))
                print_arr = copy.deepcopy(temp_arr)
                arr_shape.append(print_arr)
    
        # 두번째 줄부터 마지막 줄에 대한 시행
        for i in range(1, y_size):
            for j in range(x_size):
                if not temp_arr[i-1][j]:
                    continue
                for k in range(5):
                    nx = j + dx[k]
                    ny = i + dy[k]
                    if 0 <= nx and nx < x_size  and 0 <= ny and ny < y_size:
                        temp_arr[ny][nx] = not temp_arr[ny][nx]
                root.append((j+1, i+1))
                print_arr = copy.deepcopy(temp_arr)
                arr_shape.append(print_arr)
    
        # 마지막 줄에 대해서 모두 0인지를 체크
        check = True
        for i in range(x_size):
            if temp_arr[y_size-1][i] == True:
                check = False
        
        # 모두 0이라면 저장한 좌표값과 중간 과정들을 출력 후 끝
        if check:
            print("-------Start-------")
            show_arr(arr_shape[0])
            print("*" * (x_size*3))
            for j in range(len(root)):
                print(root[j])
                show_arr(arr_shape[j+1])
                print("*" * (x_size*3))
            print("-------Finish-------")
            break
```
